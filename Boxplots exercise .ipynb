```python
import numpy as np
from numpy.random import randn
import pandas as pd
# Import the stats library from numpy
from scipy import stats
# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#1. Load the "diamonds" dataset from Seaborn
df = sns.load_dataset("diamonds")
df
```

```python
#2. Create a boxplot of the "price" variable in the "diamonds" dataset. Set the orientation to horizontal.
sns.boxplot(x=df["price"],orient="h")
```

```python
#3. Create a boxplot of the "price" variable in the "diamonds" dataset, grouped by the "cut" variable
sns.boxplot(x="cut", y="price", data=df,linewidth=1.5)

```

```python
#4. Create a boxplot of the "price" variable in the "diamonds" dataset, grouped by both the "cut" and "color" variables.
s=sns.boxplot(y="price", hue="color", x="cut",data=df, palette="Set3")
sns.move_legend(s,"upper left",bbox_to_anchor=(1,1))
plt.show()
```

```python
#5. Create a boxplot of the "price" variable in the "diamonds" dataset, grouped by the "cut" variable and with notched boxes.
sns.boxplot(x="cut", y="price", data=df,notch=True)
```
