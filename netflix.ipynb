# Objective 
"This data frame provides a snapshot of a sample Netflix userbase, showcasing various aspects of user subscriptions, revenue and account details. Each row represents a unique user, identified by their User ID. The dataset includes information such as the user's subscription type (Basic, Standard, or Premium), the monthly revenue generated from their subscription, the date they joined Netflix (Join Date), the date of their last payment (Last Payment Date), and the country in which they are located.

Additional columns have been included to provide insights into user behavior and preferences. These columns include Device Type (e.g., Smart TV, Mobile, Desktop, Tablet),The dataset serves as a synthetic representation and does not reflect actual Netflix user data. It can be used for analysis and modeling to understand user trends, preferences, and revenue generation within a hypothetical Netflix userbase."

You can find the dataset online in kaggle using the link below https://www.kaggle.com/datasets/arnavsmayan/netflix-userbase-dataset

# Visualizing Data
To understand the relationship between data in dataset, graphing is one of the ways that used to get 
a visualized idea about that relationship.
There are many types of graphs which can be used for plotting purposes, in this case we went with using the bar chart as it recommneded for categorical data type.



```python
#Command below uses seaborn library(sns) method to generate auto count numbers on the y-axis depending on the length  of x-axis data.

#A box named as lenged with colours that represents one of the columns from a dataset indicates each bar/colour represents which observation.

#The legend will automatically replace itself wherever it is suatible for it however you can change the position of legend using the sns.move_legend() command.

#plt.show() is just to display the graph for you.
country_subsType_graph=sns.countplot(data= data_frame,  hue='Country',x='Subscription Type')
sns.move_legend(country_subsType_graph,"upper left",bbox_to_anchor=(1,1))
plt.show()

```

This graph shows that the majority of users are subscribed to the 'Basic' subscription type. It maybe because the affordability of that plan. Adding to that, France takes the first place in terms of number of users subscribed to the 'Premium' subscription type. All in all, it is concluded that the 'Basic' and the 'Standard' subscription plan are most utilized comparing to the 'Premium' one. Although the 'Standard' and 'Premium' plans total users' numbers are slighly similar, however the 'Standard' users' number is greater.
