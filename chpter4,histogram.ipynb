```python
import numpy as np
from numpy.random import randn
import pandas as pd
# Import the stats librayr from numpy
from scipy import stats
# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
# Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
dataset1= randn(100)

plt.hist(dataset1)
```

```python
dataset2=randn(80)
plt.hist(dataset2,color='red',edgecolor='black')
```

```python
plt.hist(dataset1,density=True,color='red',alpha=0.5,bins=20)
plt.hist(dataset2,density=True,alpha=0.35,bins=20)
```

```python
data1= randn(1000)
data2= randn(1000)

sns.jointplot(data1,data2)
```

```python
p=sns.jointplot(data1,data2,kind='kde')
p.set_axis_labels("Data1", "Data2")
```

```python
plt.hist(x=data1, bins=10, color='#0504aa',alpha=0.5, rwidth=0.8)
plt.grid(axis='y', alpha=0.75) #'both'
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('My Very Own Histogram')
plt.savefig("hist.jpeg",transparent=True)
```
