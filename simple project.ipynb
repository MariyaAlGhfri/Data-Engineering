```python
data_frame=pd.read_csv("Salary_data.csv")
data_frame
```

```python
sns.set_palette("Set2")
sns.scatterplot(x='YearsExperience', y='Salary', data=data_frame)
plt.title('Salary data-linear regression')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
```
